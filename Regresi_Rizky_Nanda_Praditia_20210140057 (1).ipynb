```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/CarPrice_Assignment.csv')
dataset
```

```python
%matplotlib inline
plt.scatter(dataset.horsepower,dataset.price)
plt.xlabel('horsepower')
plt.ylabel('Price')
plt.show
```

```python
reg = linear_model.LinearRegression()
reg.fit(dataset[['horsepower']],dataset.price)
```

```python
from sklearn.metrics import mean_squared_error
prediction = reg.predict(dataset[['horsepower']])

mse = mean_squared_error(dataset.price,prediction)
print("Mean Squared Error (MSE):",mse)
```
